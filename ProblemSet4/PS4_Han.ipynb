```python
#Write a loop to generate results for each year
data = {}
x = {}
olsResults = {}

for i in [1971, 1980, 1990, 2000]:
    data[i] = mod_data[mod_data.year == i]
    x[i] = data[i][['hyrsed','age','AgeSq','black','Other']]
    olsResults[i] = sm.OLS(data[i]['logwage'], x[i], missing = 'drop').fit()
    print('Parameters:', olsResults[i].params, 'for', i)
```

### Yay!! The OLS results tell us that the coefficient for years of education (beta 1) increases over the years. Next, I will use a minimizer to try to get the same results.

```python
#Before starting MLE, we have to assume the error term is normally distributed so that it is comparable to OLS
#Essentially, MLE estimator is trying to minimize the errors

#Based on theory, years of education and age should have positive impact on wage, and being black has negative impact
#Therefore, the following bounds

bnds = ((None, None), (0, None), (0, None), (None, None), (None, 0), (None, None))
```

```python
#Write a similar loop as the OLS one

data_MLE = {}
dv = {}
min = {}

for n in [1971, 1980, 1990, 2000]:
    data_MLE[n] = mod_data[mod_data.year == n]
    dv[n] = data_MLE[n]['logwage']
    #First define the objective function - based on mathematical theory, max LL = min SSE
    def objective(parameters):
        b = parameters[0]
        beta1 = parameters[1]
        beta2 = parameters[2]
        beta3 = parameters[3]
        beta4 = parameters[4]
        beta5 = parameters[5]
        #predict the DV
        yhat = b + beta1*data_MLE[n]['hyrsed'] + beta2*data_MLE[n]['age'] + beta3*data_MLE[n]['AgeSq'] + beta4*data_MLE[n]['black'] + beta5*data_MLE[n]['Other']
        #calculate the sum of squared error
        e2 = np.array((dv[n] - yhat)**2).sum()
        return e2
    #Call the minimizer
    min[n] = opt.minimize(objective, np.array([0, 0.05, 0.05, 0, -0.05, 0]), method='L-BFGS-B', bounds = bnds)
    print('Beta1 is', min[n].x[1], 'for', n)
```
